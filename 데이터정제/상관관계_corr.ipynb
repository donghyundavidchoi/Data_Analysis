```python
import pandas as pd
import numpy as np
```

# corr를 사용하여 상관관계 구하기

```python
df_corr = data.corr()
```

```python
df_corr
```

---------

# 가장 큰 상관관계와 가장 작은 상관관계 구하기
자기 자신과의 상관관계는 항상 1로 가장 큼으로 제한 조건에 1 미만을 설정   
상관관계의 크기는 부호를 빼고 생각해야하므로 abs를 통해 절대값으로 변형 후 최대값과 최소값 도출

```python
max_corr = abs(df_corr['quality'][df_corr['quality'] < 1]).max()
print(max_corr)
min_corr = abs(df_corr['quality'][df_corr['quality'] < 1]).min()
print(min_corr)
```

--------

# 가장 큰 상관관계와 가장 작은 상관관계의 합을 더하여 소수점 둘째자리까지 반올림하기

```python
print(round(max_corr + min_corr, 2))
```
