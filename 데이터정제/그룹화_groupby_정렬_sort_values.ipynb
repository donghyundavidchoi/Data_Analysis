```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/covid-vaccination-vs-death_ratio.csv')
```

```python
data
```

```python
data.isnull().sum()
```

--------

# groupby를 활용하여 국가별 데이터 구하기
# 해당 데이터는 날짜별 누적 데이터를 모두 포함하는 데이터이므로 .max()를 통해 마지막 기록만 산출

```python
data = data.groupby('country').max()
```

```python
data
```

----------

# sort_values를 활용하여 ratio를 내림차순으로 정렬

```python
data = data.sort_values(by='ratio', ascending=False)
```

```python
data
```

지브롤터 데이터가 문제가 있어 보이나 별도 확인 결과 지브롤터의 다른 데이터들도 비슷한 수치로 기록이 되어 있는 것을 보아 제대로 입력된 값으로 판단.  

-----------

# slicing을 통해 정렬된 데이터의 상위 10개와 하위 10개를 추출하고 평균을 구하기   
# head와 tail을 활용할 수도 있음

```python
upper_10_mean = data['ratio'][:10].mean()
print(upper_10_mean)
lower_10_mean = data['ratio'][-10:].mean()
print(lower_10_mean)
```

```python
upper_10_mean = data['ratio'].head(10).mean()
print(upper_10_mean)
lower_10_mean = data['ratio'].tail(10).mean()
print(lower_10_mean)
```

-------

# ratio 상위 10개 값 평균과 하위 10개 값 평균의 차이 구하기

```python
print(abs(upper_10_mean - lower_10_mean))
```
