# 이상치 삭제

# qcut()을 사용하여 동일한 수만큼 age 구간 나누기
동일한 수 대신 동일한 구간으로 나누려면 cut() 사용

```python
df1['age_range'] = pd.qcut(df1['age'], 3, labels=['young', 'adult', 'old'])
```

```python
df1
```

-------

# 각 나이 구간의 중앙값 더하기

```python
print(df1['age'][df1['age_range'] == 'young'].median() + df1['age'][df1['age_range'] == 'adult'].median() + df1['age'][df1['age_range'] == 'old'].median())
```
