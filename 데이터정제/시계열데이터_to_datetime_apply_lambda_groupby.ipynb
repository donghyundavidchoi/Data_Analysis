```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/basic2.csv')
```

```python
df
```

-------

# to_datetime을 활용하여 Date의 데이터 타입 변경

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

------

# Date로부터 날짜 정보 추출

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df
```

--------

# Events가 1일 경우 Sales에 80% 적용

```python
df['Sales_adj'] = df.apply(lambda x : x['Sales'] * 0.8 if x['Events'] == 1 else x['Sales'], axis=1)
```

```python
df
```

----------

# groupby를 사용하여 Year와 Month별 합 구하기

```python
df_groupby = df.groupby(['Year', 'Month']).sum()
```

```python
df_groupby
```

--------

# 2022년에서 가장 매출이 높은 달과 2023년에서 가장 매출이 많은 달을 추출하여 차이를 계산하고 반올림

```python
max_2022 = df_groupby['Sales_adj'].loc[2022].max()
max_2023 = df_groupby['Sales_adj'].loc[2023].max()
print(round(abs(max_2022 - max_2023)))
```
