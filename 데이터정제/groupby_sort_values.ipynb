```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/basic1.csv')
```

```python
df
```

----------

# city와 f4 대해 groupby실행하여 f5의 평균값 도출

```python
df_groupby = df.groupby(['city', 'f4'])[['f5']].mean()
```

```python
df_groupby
```

--------

# f5 값에 대하여 내림차순으로 정렬

```python
df_groupby = df_groupby.sort_values(by='f5', ascending=False)
```

```python
df_groupby
```

-------

# f5 평균 상위 7개 값을 합하여 소수점 두자리 수로 반올림

```python
print(round(df_groupby['f5'][0:7].sum(), 2))
```
